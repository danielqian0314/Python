```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
s = np.random.poisson(0.5, 10000)
plt.hist(s), plt.title("poisson distribution of 10000 variables with lambda =0.5")
```

```python
def p(n):
    p=(n-1)/(2*n*n)
    return p
s = np.random.binomial(10, p(10),10000)
plt.hist(s), plt.title("binomial distribution of 10000 variables with n =10")
```

```python
s = np.random.binomial(100, p(100),10000)
plt.hist(s), plt.title("binomial distribution of 10000 variables with n =100")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.data
from scipy import signal

def makeGaussian(c):
    array = np.zeros(shape=(401,401))
    for x in range(401):
        for y in range(401):
            array[x][y]=1/(2*np.pi*pow(c, 2))*np.exp(-(pow(x-201, 2.0)+pow(y-201, 2))/(2*pow(c, 2)))                                                     
    return array                                                   
```

```python
plt.imshow(makeGaussian(1)), plt.title("gaussian with covariance=1")
```

```python
plt.imshow(makeGaussian(10)), plt.title("gaussian with covariance=10")
```

```python
plt.imshow(makeGaussian(100)), plt.title("gaussian with covariance=100")
```

```python
fgaussian1=np.fft.fft2(makeGaussian(1))
fgaussian1=np.fft.fftshift(fgaussian1)
fgaussian1img=np.abs(fgaussian1)
plt.imshow(fgaussian1img), plt.title("FT of gaussian with covariance=1")
```

```python
fgaussian10=np.fft.fft2(makeGaussian(10))
fgaussian10=np.fft.fftshift(fgaussian10)
fgaussian10img=np.abs(fgaussian10)
plt.imshow(fgaussian10img), plt.title("FT of gaussian with covariance=10")
```

```python
fgaussian100=np.fft.fft2(makeGaussian(100))
fgaussian100=np.fft.fftshift(fgaussian100)
fgaussian100img=np.abs(fgaussian100)
plt.imshow(fgaussian100img), plt.title("FT of gaussian with covariance=100")
```

```python
ffgaussian100=np.fft.fft2(fgaussian100)
ffgaussian100img=np.abs(ffgaussian100)
plt.imshow(ffgaussian100img), plt.title("FT of FT of gaussian with covariance=100")
```

```python
cgaussian100=signal.fftconvolve(makeGaussian(100),makeGaussian(100))
plt.imshow(cgaussian100), plt.title("convolution of two gaussian with covariance=100")
    
```

```python
def makeTriangular(a,b):
    array = np.zeros(shape=(100,100))
    for x in range(100):
        for y in range(100):
            if np.abs(x-50)<a and np.abs(y-50)<b:
                array[x][y]=(1-np.abs(x-50)/a)*(1-np.abs(y-50)/b)                                                     
    return array 


plt.imshow(makeTriangular(25,25)), plt.title("2d triangular function with width=50")
```

```python
fttriangular50=np.fft.fft2(makeTriangular(25,25))
fttriangular50=np.fft.fftshift(fttriangular50)
fttriangular50img=np.abs(fttriangular50)
plt.imshow(fttriangular50img), plt.title("FT of 2d triangular function with width=50")
```

```python
def makeRectangular(a,b):
    array = np.zeros(shape=(100,100))
    for x in range(100):
        for y in range(100):
            if np.abs(x-50)<a and np.abs(y-50)<b:
                array[x][y]=1                                                     
    return array 

plt.imshow(makeRectangular(25,25)), plt.title("2d rectangular function with width=50")
```

```python
ftrectangular50=np.fft.fft2(makeRectangular(25,25))
ftrectangular50=np.fft.fftshift(ftrectangular50)
ftrectangular50img=np.abs(ftrectangular50)
plt.imshow(ftrectangular50img), plt.title("FT of 2d rectangular function with width=50")
```

```python
ft2rectangular50=np.fft.fft2(ftrectangular50)
ft2rectangular50img=np.abs(ft2rectangular50)
plt.imshow(ft2rectangular50img), plt.title("FT2 of 2d rectangular function with width=50")
```

```python
ftmft2rectangular50=ftrectangular50*ft2rectangular50
ftmft2rectangular50img=np.abs(ftmft2rectangular50)
plt.imshow(ftmft2rectangular50img), plt.title("mutiply FT and FT2 of 2d rectangular function with width=50")
```

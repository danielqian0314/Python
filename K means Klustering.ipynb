```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.sparse.linalg as sla

#a)Load all images in data Matrix X
X=np.zeros((135,38880))
Type=['centerlight', 'glasses', 'happy','noglasses', 'normal', 'sad','sleepy', 'surprised','wink']
for i in range(15):
    for j in range(9):
        filename="yalefaces_cropBackground\subject{:0>2d}.{}".format(i+1,Type[j])
        X[i*9+j]=plt.imread(filename)[:,:,0].reshape((38880,))
```

```python
error=np.zeros((14,))
K=4
N=X.shape[0]
centers=np.zeros((K,38880))
for k in range(K):
    centers[k]=X[np.random.randint(135)]    

for iterator in range(10):   
    gamma=np.zeros((N,K))
    for n in range(N):  
        k=np.argmin(np.linalg.norm(X[n]-centers,axis=1)*np.linalg.norm(X[n]-centers,axis=1))
        gamma[n][k]=1

    for k in range(K):
        sumgammax=0
        sumgamma=0
        for n in range(N):        
            sumgammax+=gamma[n][k]*X[n]
            sumgamma+=gamma[n][k]
        centers[k]=sumgammax/sumgamma

for n in range(N):
    for k in range(K):
        error[K-2]+=(np.linalg.norm(X[n]-centers[k])*np.linalg.norm(X[n]-centers[k]))*gamma[n][k]

print('Errors:',error[K-2])
for i in range(K):
    plt.subplot(1,K,i+1)
    plt.imshow(centers[i].reshape((243,160)),cmap='gray'),plt.title(i+1)
```

```python
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(X[np.argmax(gamma[:,i])].reshape((243,160)),cmap='gray'),plt.title('k={}'.format(i+1))
```
